```python
# import libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
import datetime
import smtplib # for sending emails
```

```python
# connect to website
url = 'https://www.amazon.in/Allen-Solly-Regular-T-Shirt-ASKPQRGF701338_Medium/dp/B08KTVFFHF/?_encoding=UTF8&ref_=pd_hp_d_btf_a2i_gw_cml'

headers = {
  "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36","Accept-Encoding": "gzip, deflate, br, zstd","Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8","DNT":"1","Connection":"close","Upgrade-Insecure-Requests": "1"
}

page = requests.get(url, headers=headers)

Soup1 = BeautifulSoup(page.content, "html.parser")

Soup2 = BeautifulSoup(Soup1.prettify(), "html.parser")

title = Soup2.find('span',id='productTitle').get_text()
price = Soup2.find('span', class_='a-price-whole').get_text()

print(title)
print(price)
```

```python
# Clean up the data a little bit
price = price.strip()
title = title.strip()
```

```python
# Create a Timestamp for your output to track when data was collected
import datetime

today = datetime.date.today()

print(today)
```

```python
# Create CSV and write headers and data into the file

import csv

header = ['Title', 'Price', 'Date']
data = [title, price, today]


with open('AmazonWebScraperDataset.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```
